```python
def iou(rect1, rect2):
    xmin1, ymin1, xmax1, ymax1 = rect1
    xmin2, ymin2, xmax2, ymax2 = rect2
    s1 = (xmax1 - xmin1) * (ymax1 - ymin1)
    s2 = (xmax2 - xmin2) * (ymax2 - ymin2)

    sum_area = s1 + s2

    left = max(xmin2, xmin1)
    right = min(xmax2, xmax1)
    top = max(ymin2, ymin1)
    bottom = min(ymax2, ymax1)

    if left >= right or top >= bottom:
        return 0

    intersection = (right - left) * (bottom - top)
    return intersection / (sum_area - intersection ) * 1.0
```

```python
import torch
import pandas as pd
dataset = "../../ear_detection1/dataset/eval_set"
dataset_json = "../../ear_detection1/dataset/eval_set_json/bbox.json"
ear_eavluation = "../../EAR_EVALUATION"

imgs = os.listdir(dataset)

with open(dataset_json, "r") as bbox_file:
    gt_bbox = json.load(bbox_file)


df_total = pd.DataFrame()
for img in imgs:
    image_path = os.path.join(dataset, img)
    img_index = os.path.splitext(img)[0]
    image = mmcv.imread(image_path, channel_order='rgb')
    result = inference_detector(model, image)
    
    
    gt = gt_bbox[img_index]
    gt[2] = gt[0] + gt[2]
    gt[3] = gt[1] + gt[3]
    
    bbox_count = 0
    for boundingbox in result.pred_instances.bboxes:
        if bbox_count == 0:
            pred = boundingbox.clone().type(torch.int)
            pred = pred.tolist()

            df_temp = pd.DataFrame(index = [bbox_count],columns = ['Object name','IoU','Confidence'])
            df_temp.at[bbox_count,'Object name'] = img
            df_temp.at[bbox_count,'IoU'] = iou(gt,pred)
            df_temp.at[bbox_count, 'Confidence'] = result.pred_instances.scores[bbox_count].cpu().numpy()

            df_total = pd.concat([df_total, df_temp], axis=0,ignore_index=True)
            bbox_count+=1
df_total.to_csv(os.path.join(ear_eavluation,"ear.csv"))
print(df_total)
```

```python

import openpyxl
ear_eavluation = "../../EAR_EVALUATION"
csv_files = os.listdir(ear_eavluation)
csv_files.pop(3)
df_final = pd.DataFrame()
for csv_file in csv_files:
    csv_path = os.path.join(ear_eavluation, csv_file)
    df = pd.read_csv(csv_path, index_col = 0)
    df_final = pd.concat([df_final, df],axis=0, ignore_index=True)

    
df_final = df_final.sort_values(by=['Confidence'],ascending=False)
df_final = df_final.reset_index(drop=True)


tp = 0
fp = 0

recall = 0
precision = 0
recall_list = []
precision_list = []

for i in range(0, len(df_final)):

    if float(df_final.at[i, "IoU"]) >= 0.5:
        tp += 1
    else:
        fp += 1
    
    recall = tp/720
    precision = tp/(tp+fp)
    

    precision_list.append(precision)

    recall_list.append(recall)
       

col_name = df_final.columns.tolist()
col_name.insert(col_name.index('Confidence')+1,'Precision')
col_name.insert(col_name.index('Precision')+1,'Recall')
    
df_final=df_final.reindex(columns=col_name)
    
df_final['Precision']=precision_list
df_final['Recall']=recall_list
df_final.to_excel(os.path.join(ear_eavluation,"precision_recall.xlsx"))
```
