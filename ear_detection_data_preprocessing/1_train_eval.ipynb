```python
data = "../data"
bbox_json = "../bbox_json"
train_set = "../dataset/train_set"
eval_set = "../dataset/eval_set"
train_set_json = "../dataset/train_set_json"
eval_set_json = "../dataset/eval_set_json"

if not os.path.isdir(train_set):
    os.makedirs(train_set)
if not os.path.isdir(eval_set):
    os.makedirs(eval_set)
if not os.path.isdir(train_set_json):
    os.makedirs(train_set_json)
if not os.path.isdir(eval_set_json):
    os.makedirs(eval_set_json)
    
with open("../bbox_json/bbox.json", 'r') as bbox_file:
    bbox = json.load(bbox_file)
imgs = os.listdir(data)
imgs = sorted(imgs, key= lambda s : int(os.path.splitext(os.path.basename(s))[0]), reverse = False)

train_bbox = {}
eval_bbox = {}

train_index = 0
eval_index = 0
for i in range(0, len(imgs)):
    if ( i >= 900 and i < 1080 ) or ( i >= 1440 and i < 1620 ):
        src = os.path.join(data, imgs[i])
        dst = os.path.join(eval_set, "{index}.png".format(index =eval_index))
        shutil.copyfile(src, dst)
        
        eval_bbox[eval_index] = bbox[str(i)]
        
        eval_index += 1
    else:
        src = os.path.join(data, imgs[i])
        dst = os.path.join(train_set, "{index}.png".format(index =train_index))
        shutil.copyfile(src, dst)
        
        train_bbox[train_index] = bbox[str(i)]
        print(bbox[str(i)])
        
        train_index += 1


with open(os.path.join(train_set_json, "bbox.json"), 'w') as train_json_file:
    json.dump(train_bbox, train_json_file, indent=4)
with open(os.path.join(eval_set_json, "bbox.json"), 'w') as eval_json_file:
    json.dump(eval_bbox, eval_json_file, indent=4)

```
