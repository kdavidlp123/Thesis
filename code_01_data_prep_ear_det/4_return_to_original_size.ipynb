```python
import os
import json
import glob
import shutil
import cv2
```

```python
degrees = ['15cm_0mm_0deg', '15cm_25mm_5deg', '15cm_50mm_10deg', '20cm_0mm_0deg', '20cm_25mm_5deg', '20cm_50mm_10deg']
```

# 還原照片

```python
videos = "../0_original_video"
crop_512_center_json = "../1_crop_512_center_json"
after_inpainting = "../4_after_inpainting"
return_to_ori_size = "../4_return_to_ori_size"
if not os.path.isdir(return_to_ori_size):
    os.makedirs(return_to_ori_size)

for name in os.listdir(videos):
    crop_512_center_json_name = os.path.join(crop_512_center_json, name)
    videos_name = os.path.join(videos, name)
    after_inpainting_name = os.path.join(after_inpainting, name)
    return_to_ori_size_name = os.path.join(return_to_ori_size, name)
    
    for deg in degrees:
        count = 0
        
        crop_512_center_json_name_deg = os.path.join(crop_512_center_json_name, deg)
        videos_name_deg = os.path.join(videos_name, deg)
        after_inpainting_name_deg = os.path.join(after_inpainting_name, deg)
        return_to_ori_size_name_deg = os.path.join(return_to_ori_size_name, deg)
        
        with open(os.path.join(crop_512_center_json_name_deg, "512_size_center.json"), "r") as f:
            center = json.load(f)
        
        if not os.path.isdir(return_to_ori_size_name_deg):
            os.makedirs(return_to_ori_size_name_deg)
        
        video = os.path.join(videos_name_deg, "resize.mp4")
        cap = cv2.VideoCapture(video)
        if not cap.isOpened():
            print("Cannot open camera")
            exit()

        while True:
            ret, frame = cap.read()
            if not ret:
                print("Cannot receive frame")
                break
            
            
            repaired_512_img = cv2.imread(os.path.join(after_inpainting_name_deg, "test{index}.png".format(index = count)))
            frame[center[1] - 256 : center[1] + 256, center[0] - 256 : center[0] + 256] = repaired_512_img
            
            
        
            cv2.imwrite(os.path.join(return_to_ori_size_name_deg, '{index}.png'.format(index = count)), frame)
            count = count + 1
            
            
        cap.release()
        cv2.destroyAllWindows()
```
