```python
import cv2
import numpy as np
import pandas as pd
import os
import glob
import json
import copy
```

```python
degrees = ['15cm_0mm_0deg', '15cm_25mm_5deg', '15cm_50mm_10deg', '20cm_0mm_0deg', '20cm_25mm_5deg', '20cm_50mm_10deg']
```

```python

original_video = "../0_original_video"
crop_512_center_json = "../1_crop_512_center_json"
jsons = "../2_json"
crop_512_for_inpainting = "../3_crop_512_for_inpainting"
masks_for_inpainting = "../3_mask_for_inpainting"


names = os.listdir(original_video)    

for name in names:
    jsons_name = os.path.join(jsons, name)
    original_video_name= os.path.join(original_video, name)
    crop_512_center_json_name = os.path.join(crop_512_center_json, name)
    crop_512_for_inpainting_name = os.path.join(crop_512_for_inpainting, name)
    masks_for_inpainting_name = os.path.join(masks_for_inpainting, name)

    for deg in degrees:
        
        original_video_name_deg = os.path.join(original_video_name, deg)
        jsons_name_deg = os.path.join(jsons_name, deg)
        crop_512_center_json_name_deg = os.path.join(crop_512_center_json_name, deg)
        crop_512_for_inpainting_name_deg = os.path.join(crop_512_for_inpainting_name, deg)
        masks_for_inpainting_name_deg = os.path.join(masks_for_inpainting_name, deg)
        
        if not os.path.isdir(crop_512_for_inpainting_name_deg):
            os.makedirs(crop_512_for_inpainting_name_deg)
        if not os.path.isdir(masks_for_inpainting_name_deg):
            os.makedirs(masks_for_inpainting_name_deg) 
         
        count = 0

        if os.path.exists(os.path.join(jsons_name_deg, "keypoint_location.json")):

            with open(os.path.join(jsons_name_deg, "mask.json"), "r") as f_mask:
                mask = json.load(f_mask)
            with open(os.path.join(jsons_name_deg, "visible.json"), "r") as f_visible:
                visible = json.load(f_visible)
            with open(os.path.join(jsons_name_deg, "keypoint_location.json"), "r") as f_keypoint_location:
                keypoint_location = json.load(f_keypoint_location)
            with open(os.path.join(crop_512_center_json_name_deg, "512_size_center.json"), "r") as f_center_file:
                center_of_512 = json.load(f_center_file)





            video = os.path.join(original_video_name_deg, "resize.mp4")
            cap = cv2.VideoCapture(video)
            if not cap.isOpened():
                print("Cannot open camera")
                exit()

            while True:
                ret, frame = cap.read()
                if not ret:
                    print("Cannot receive frame")
                    break

                frame_512 = frame[center_of_512[1] - 256 : center_of_512[1] + 256, center_of_512[0] - 256 : center_of_512[0] + 256]
                mask_ori_size = 255 - np.zeros((frame.shape[0], frame.shape[1]), np.uint8)
                for i in range(0,len(mask["frame_0"])):
                    imgRoi = frame[mask["frame_"+str(count)][str(i)]['left_top'][1] : mask["frame_"+str(count)][str(i)]['right_bot'][1], mask["frame_"+str(count)][str(i)]['left_top'][0] : mask["frame_"+str(count)][str(i)]['right_bot'][0]]  # imgRoi = frame[p1y:p2y,p1x:p2x]
                    imgRoi_mask = np.zeros((imgRoi.shape[0], imgRoi.shape[1]), np.uint8)

                    mask_ori_size[mask["frame_"+str(count)][str(i)]['left_top'][1] : mask["frame_"+str(count)][str(i)]['right_bot'][1], mask["frame_"+str(count)][str(i)]['left_top'][0] : mask["frame_"+str(count)][str(i)]['right_bot'][0]] = imgRoi_mask
                mask_512 = mask_ori_size[center_of_512[1] - 256 : center_of_512[1] + 256, center_of_512[0] - 256 : center_of_512[0] + 256]



                cv2.imwrite(os.path.join(crop_512_for_inpainting_name_deg, 'test{index}.png'.format(name_0 = name, index = str(count))), frame_512)
                cv2.imwrite(os.path.join(masks_for_inpainting_name_deg, 'mask{index}.png'.format(name_0 = name, index = str(count))), mask_512)
                count = count + 1


            cap.release()
            cv2.destroyAllWindows()
```
