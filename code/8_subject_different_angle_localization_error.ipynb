```python
mmpose_result="../../../mmpose/result"
names = os.listdir(mmpose_result)


final = "../7_final"



        


final_mean = []
final_std = []
for deg in degrees:
    subject_mean = []
    subject_std = []
    for l in range(0, 4):
        subject_mean.append([])
        subject_std.append([])
    
    for name in names:

        

        





        mmpose_result_name_localization_error_deg = os.path.join(mmpose_result, name, "localization_error", deg)
        file_num = len(os.listdir(mmpose_result_name_localization_error_deg))
        interval = file_num / 4
        x = np.arange(0, file_num, interval)
        x = [round(num) for num in x]


        if len(x) > 4:
            x.pop(4)

        
        
        
        for i in range(0, len(x)):
            file = os.path.join(mmpose_result_name_localization_error_deg, "test_{index}.json".format(index = x[i]))
        
            with open(file,"r") as f:
                error_f = json.load(f)
            
            
            acupoints = []
            
            for j in range(0,len(error_f)):
                if error_f[str(j)]:
                    acupoints.append(error_f[str(j)][0])

            subject_mean[i].append(np.mean(np.array(acupoints))*10)
            subject_std[i].append(np.std(np.array(acupoints))*10)

    final_mean.append(subject_mean)
    final_std.append(subject_std)
print(final_mean)
# print(len(final_mean[0]))
# plt.figure(figsize=(15,10))
for i in range(0, len(degrees)):
    plt.figure(figsize=(15,10))
    for j in range(0 ,len(final_mean[i])):
        print(final_mean[i][j])
        plt.bar(np.arange(8) + j*0.2, final_mean[i][j], yerr=final_std[i][j], align='center', label = '{index} degrees'.format(index = 0 + 10*j), ecolor='black', capsize=5, width=0.2)
        
#     plt.xlabel('acpuoint index' , fontsize = 30)
    plt.ylabel('mm' , fontsize = 30)

    plt.xticks(np.arange(8) + 0.3, ["subject_0", "subject_1", "subject_2", "subject_3", "subject_4", "subject_5", "subject_6", "subject_7"], fontsize=25, rotation = -30)


    plt.yticks(fontsize=30)
    plt.title('Localization error', fontsize = 30)

    plt.legend(fontsize=20)

    plt.tight_layout()
    plt.savefig(os.path.join(final, degrees[i],'{index}.png').format(index = 1))
    plt.show()
```

```python
final = "../7_final"
width=0.25

for deg in degrees:
    json_file = os.path.join(final, deg, "different_angle.json")

    with open(json_file, "r") as f:
        result = json.load(f)
    

    means = []
    stds = []
    
    for j in range(0 ,len(result)):
        mean = []
        std = []

        for k in range(0 , 21):

            acupoint_errors = np.array(result[str(j)][str(k)])
            mean.append(np.mean(acupoint_errors)*10.00)
            std.append(np.std(acupoint_errors)*10.00)


        means.append(mean)
        stds.append(std)

    plt.figure(figsize=(15,10))

    for l in range(0 ,len(means)):

        plt.bar(np.arange(21) + l*0.2, means[l], yerr=stds[l], align='center', label = '{index} degrees'.format(index = 0 + 10*l), ecolor='black', capsize=3, width=0.2)
    
    plt.xlabel('acpuoint index' , fontsize = 30)
    plt.ylabel('mm' , fontsize = 30)

    plt.xticks(np.arange(21) + 0.3, [str(x) for x in list(range(0 , 21))], fontsize=30)
    
    
    plt.yticks(fontsize=30)
    plt.title('Localization error', fontsize = 30)

    plt.legend(fontsize=20)

    plt.tight_layout()
    plt.savefig(os.path.join(final, deg,'{index}.png').format(index = 0))
    plt.show()
        
        
```
