```python
import json
import copy
import os
import cv2
```

```python
tracking_visualization = "../tracking_visualization"
original_video = "../0_original_video"
jsonfolder = "../2_json"
names = os.listdir(original_video)
```

```python
degrees = ['15cm_0mm_0deg']
```

```python
for name in names:
    tracking_visualization_name_image = os.path.join(tracking_visualization, name,"image")
    tracking_visualization_name_image_with_tracking_box = os.path.join(tracking_visualization, name,"image_with_tracking_box")
    tracking_visualization_name_image_with_point = os.path.join(tracking_visualization, name,"image_with_point")
    
    if not os.path.isdir(tracking_visualization_name_image):
        os.makedirs(tracking_visualization_name_image)
    if not os.path.isdir(tracking_visualization_name_image_with_tracking_box):
        os.makedirs(tracking_visualization_name_image_with_tracking_box) 
    if not os.path.isdir(tracking_visualization_name_image_with_point):
        os.makedirs(tracking_visualization_name_image_with_point)   
        
        
        
    for deg in degrees:
        count = 0
        mask_json= os.path.join(jsonfolder,name,deg,"mask.json")
        keypoint_location_json = os.path.join(jsonfolder,name,deg,"keypoint_location.json")
        
        with open(os.path.join(mask_json), "r") as mask_f:
            box_file = json.load(mask_f)
        with open(os.path.join(keypoint_location_json), "r") as keypoint_f:
            point_file = json.load(keypoint_f)
        print(point_file)
        video = os.path.join(original_video,name,deg,"resize.mp4")
        cap = cv2.VideoCapture(video)
        if not cap.isOpened():
            print("Cannot open camera")
            exit()
        while True:
            ret, frame = cap.read()
            
            frame_box = frame.copy()
            frame_point = frame.copy()
            
            boxes = box_file["frame_"+str(count)]
            points = point_file["frame_"+str(count)]
            
            for i in range(0,len(points)):
                left_up = (int(boxes[str(i)]["left_top"][0]), int(boxes[str(i)]["left_top"][1]))
                right_down = (int(boxes[str(i)]["right_bot"][0]), int(boxes[str(i)]["right_bot"][1]))
                cv2.rectangle(frame_box, left_up, right_down,(0,0,255), 3) 
                                  
                point = (int(points[str(i)]["keypoint_location"][0]), int(points[str(i)]["keypoint_location"][1]))
                cv2.circle(frame_point, point, 3, (0,0,255), -1)
            
            
            if not ret:
                print("Cannot receive frame")
                break
            keyName = cv2.waitKey(1)
            if keyName == ord('q'):
                    break
            cv2.imwrite(os.path.join(tracking_visualization_name_image, str("%04d"%count)+".png"), frame)
            cv2.imwrite(os.path.join(tracking_visualization_name_image_with_tracking_box, str("%04d"%count)+".png"), frame_box)
            cv2.imwrite(os.path.join(tracking_visualization_name_image_with_point, str("%04d"%count)+".png"), frame_point)
            cv2.imshow('oxxostudio', frame)
            count+=1
            
        cap.release()
        cv2.destroyAllWindows()
```
